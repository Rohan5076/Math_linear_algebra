# **Probability**
Probability is a numerical concept used to measure the chance of any specific event or outcome occurring. The value of the probability ranges from 0 to 1. If the value is closer to 1. Then we can assume it has a high probability to occur. In contrast, if the value lies closer to 0. Then it can be said it has a high probability to disappear. Probability is an effective technique that is used for non-trivial applications. Such as

* Spam Filter for inboxes in Gmail.
* Image Recognition System For Medical Image Analysis.
Two different processes are used to find probabilities.

##**i). Experimental or Empirical Probabilities.**

ii). Theoretical Probabilities.

➤ Experimental or Empirical Probabilities
Empirical Probabilities is the process of estimating the probability by experimenting. We consider a coin toss event; we get only two outcomes head or tail if we throw a coin. Suppose, we want to find the probabilities of getting head. We can find it by following the below steps.

i). At first, we can flip the coin several times. We assume flip the coin 200 times.

ii). We can count the number of times the coins landed on the head. We assume the coin landed on the head 79 times.

iii). Now, we can divide the number heads by the numbers of the total of times we flip the coins.

So the probability of getting head = 79/200 = 0.395

To illustrate, we can consider another real-life example; you want to launch a new product with two types of variation in the market. You have to know the people’s preferences before launching it into the market.

To perform this experiment, you select hundreds of people randomly, and then you talk them to choose products in their preference. After completion, you find that 85 people prefer the first product and the rest choose the second one. So, the **empirical probability** of the first product is = 85/100.

##**➤ Theoretical Probabilities**
Theoretical Probabilities is the process of estimating probabilities based on reasoning. Consider the previous example of the coin toss. We have to follow the steps below to find the theoretical probabilities.

i). At first, we have to know the total number of favorable outcomes. Here, for tail, it has only one possible outcome.

ii). Finally, we have to know the total number of outcomes. Here, we may have two possible outcomes: head and tail.

So, the theoretical probabilities = 1/2= 0.5

Let’s assume another real-life example, taking a red card from the deck of cards. If we calculate the probability of getting red, it will be.

P(red) = 13/52

Here, the number of favorable outcomes of getting red is 13. And a total number of outcomes is 52.

#**✪ Some Basic Probability Terminologies**
###**➤ Outcomes & Event:**
We can denote outcomes as an event. The total number of outcomes from an experiment is called an event. We can also mention the outcome as all possible results from an event. Suppose we want to forecast the weather for the upcoming week. There might be different kinds of possible scenarios.

(i) All days of the week will be sunny, rainy, or cloudy.

(ii) Three days will be rainy, sunny, or cloudy, and another three days will be cloudy, sunny, or rainy.

Also, many more scenarios can happen.

All of the possible results individually are called outcomes. And all of the possible results on the given condition together are called events.

###**➤ Sample Space:**
Sample space denotes all the possible outcomes in a set. Suppose a rolling dice has six possible outcomes. We can write all of the outcomes in a set. Such as : {1,2,3,4,5,6}. Also, an event is a subset of Sample Space.

#**✪ Types of different Kinds of events in Probability**
In this part, we will try to explain the different kinds of events in probability.

###**➤ Impossible and sure event:**
If the possibility of occurring an event is 0, it is called an impossible event. And if the possibility of occurring any event is 1, it is called a sure event. Suppose you roll a dice that has six possible outcomes. The chance of getting a number between 1–6 is 1. It is a sure event for this experiment. But the possibility of getting a number greater than 6 is 0. Because the possible outcomes lie between 1–6, it is an impossible event, and this experiment is called an impossible event.

###**➤ Simple event:**
The event with only one possible outcome is called a Simple event. Example: s={23}, this event has only one possible outcome. So, it is a simple event.

###**➤ Compound Event:**
The event with more than one possible outcome is called a Compound Event. Example: s={23,34,45}, this event has three possible outcomes.

###**➤ Independent and Dependent Event:**
If the event is not dependent on any other event or an event completely unaffected from other events is called an independent event. And the event that is affected by other events is called Dependent events. Suppose you draw two balls from a bag one after another. If you draw the first ball and then the second ball without putting back the first ball, it will be affected by the first ball. In contrast, if you put the first ball back in the bag and draw the second ball. It will not be affected by the first ball.

###**➤ Mutually Exclusive Event:**
Mutually Exclusive Event is a thing that denotes that two things can not happen at a time. It can also indicate that the occurring of any event is completely independent of the event of occurring other. Such as throwing a coin has two events s ={H, T}, these two events are totally different

#**✪ Some rules in probability**
###**➤Additive Rules of Probability**
Additive rules state that the probability for getting A or B will be the sum of the probability of getting A and the probability of getting B. It is also denoted by ‘U’.

###**➣ Additive Rules for Mutually Exclusive Event:**
Here, two events are not affected by each other. The formula for this event is

P(A U B) = P(A) + P(B)

If the event has more than two events, then the formula becomes

P(A ∪ B C…..) = P(A) + P(B) + P(C)………..

Suppose we pick up a card from a deck of cards. Now, The probability of getting the red or black card.

P(red or black) = 26/52 + 26/52

Here, red and black events are not affecting each other.

Here, two events are affected by each other. The formula for this event is

P(A U B) = P(A) + P(B) — P(A ∩ B).

##**➤ Multiplication Rules of Probability**
When two events occur one after another here multiplication rules are used. It denotes by the ‘∩’ symbol.

##**➣ Multiplication Rule for Mutually Exclusive Event: Here, Two events occur one after another, and both are independent. The formula for this event is**
P(A ∩ B) = P(A)*P(B)

Suppose, we roll a dice two times and want to find the probability of getting 1 and 3.

P(1 ∩ 3) = P(1)*P(3)

If we roll dice each time, the probability of getting a number between 1 to 6 is 1/6. Here, the probability of the current event of getting a number in this range is not dependent on the previous event. If we roll a dice two, three, or whatever times, we get the same probability value for getting numbers between 1 and 6.

In this example,

P(1)= 1/6

P(3) = 1/6

So, P(1 ∩ 3) = 1/6 *1/6

##**➣ Multiplication Rule for Not Mutual Exclusive Event:**
Here, Two events occur one after another, and both are dependent on each other. Suppose, you are taking a card from a deck of 52 cards. You pick up a red card. Again, you take a card without putting the red card back on the deck. Here, two things may happen.

i) When you take the first card, at that time, the total number of cards is 52.

ii) Secondly, When you pick up the second card, at that time, the number of the total card is 51 as you do not put back the first item in the deck.

In this scenario, we find that the two events are dependent on each other. Suppose, you want to take two cards. Firstly, you want to pick Red card and then a King. So the probability of getting Red and King is

P(R ∩ K) = P(R)*P(K/R) ,

We find a new term, P(K/R), called conditional probability. We will describe it in the next part. It is called the Probability of K given R. That means we want to calculate the probability of K, which depends on the previous event. So,

P(R ∩ K) = 1/52 * 1/51

Here, P(R) = 1/52, As we have taken one card from the total of 52 cards.

P(K/R) = 1/51, As we have taken one card from the total of 51 cards.

#**Marginal Distribution**
In probability theory and statistics, the marginal distribution of a subset of a collection of random variables is the probability distribution of the variables contained in the subset. It gives the probabilities of various values of the variables in the subset without reference to the values of the other variables — If that was too much jargon, to put it simply, the marginal probability is the probability of an event irrespective of the outcome of another variable — P(A) or P(B).

#**Conditional Probability**
The conditional probability concept is one of the most fundamental in probability theory and in my opinion is a trickier type of probability. It defines the probability of one event occurring given that another event has occurred (by assumption, presumption, assertion or evidence).
